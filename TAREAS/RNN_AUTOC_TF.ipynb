```python
import tensorflow as tf
import numpy as np


def complete_tf(prompt, max_new_tokens=120, temperature=0.8, eos_token_id=None):
    ids = encode(prompt)
    temperature = max(temperature, 1e-8)
    
    for _ in range(max_new_tokens):
        x = np.array(ids[-block_size_tf:], dtype=np.int64)
        if x.shape[0] < block_size_tf:
            pad = np.zeros(block_size_tf - x.shape[0], dtype=np.int64)
            x = np.concatenate([pad, x])
        
        x = x.reshape(1, block_size_tf)
        logits = model_tf(x, training=False).numpy()[0, -1, :]
        logits = logits / temperature
        logits = logits - np.max(logits)
        probs = np.exp(logits)
        probs = probs / np.sum(probs)
        
        next_id = int(np.random.choice(len(probs), p=probs))
        ids.append(next_id)
        
        if eos_token_id is not None and next_id == eos_token_id:
            break
    
    return decode(ids)

# 0.75 -> 0.3; mas conservador
print(complete_tf("ef es_primo(n):", max_new_tokens=90, temperature=0.2)[:450])
```
